```python
import requests
url = 'http://dataservices.imf.org/REST/SDMX_JSON.svc/'
key = 'CompactData/IFS/A..FPOLM_PA'
data = requests.get(f'{url}{key}').json()['CompactData']['DataSet']
data
```

```python
import pandas as pd

results = []
for country_data in data['Series']:
    country_code = country_data['@REF_AREA']
    observation_data = country_data['Obs']
    for obs_datum in observation_data:
        print(obs_datum)
        year = obs_datum['@TIME_PERIOD']
        try:
            value = obs_datum['@OBS_VALUE']
        except:
            value = 'DATAERROR_WITHSTATUS_' + str(obs_datum['@OBS_STATUS'])

        results.append({
            'year':year,
            'value':value,
            'country_code':country_code,
        })

ax = pd.DataFrame(results)
```

```python
ax.to_csv('output.csv', index=False)
```
